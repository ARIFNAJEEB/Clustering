```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('crime_data.csv')
data
```

```python
data.info()
```

```python
crime= data.drop('Unnamed: 0',axis=1)
crime
```

Normalization

```python
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm = norm_func(crime.iloc[:,:])
df_norm
```

# KMeans Clustering

```python
fig=plt.figure(figsize=(10,8))
WCSS = []
for i in range(1,11):
    clf=KMeans(n_clusters=i)
    clf.fit(df_norm)
    WCSS.append(clf.inertia_)
plt.plot(range(1,11),WCSS)
plt.title('The elbow chart')
plt.xlabel('WCSS')
plt.ylabel('the number of clusters')
plt.show()
```

```python
clf=KMeans(n_clusters=5)
ykmeans = clf.fit_predict(df_norm)
```

```python
ykmeans
```

```python
clf.labels_
```

```python
clf.cluster_centers_
```

```python
clf.inertia_
```

```python
md= pd.Series(ykmeans) #converting numpy array to pandas series object
crime['clust']=md # creating a new colomn and assigning it to a new colomn
crime
```

```python
crime.groupby(crime.clust).mean()
```

```python
WCSS
```

kmeans vizualization

```python
plt.figure(figsize=(15,8))
plt.scatter(crime['clust'], data['Unnamed: 0'], c=clf.labels_, s=300, marker='*')
plt.show()
```

# DBSCAN Clustering

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
```

```python
crime
```

```python
array=crime.values
array
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
X
```

```python
dbscan = DBSCAN(eps=1.25, min_samples=5)
dbscan.fit(X)
```

```python
#noisy samples are given te labels -1
dbscan.labels_
```

```python
c= pd.DataFrame(dbscan.labels_,columns=['clusters'])
```

```python
c
```

```python
df=pd.concat([data,c],axis=1)
df
```

```python
d1 = dbscan.labels_
d1
```

```python
import sklearn
sklearn.metrics.silhouette_score(X,d1)
```

```python
from sklearn.cluster import KMeans
clf=KMeans(n_clusters=5)
y_kmeans = clf.fit_predict(X)
```

```python
y_kmeans
```

```python
cl1 = pd.DataFrame(y_kmeans,columns=['kcluster'])
cl1
```

```python
df1= pd.concat([df,cl1],axis=1)
df1
```

silhouette_score

```python
sklearn.metrics.silhouette_score(X,y_kmeans)
```

Dendogram

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
p = np.array(df_norm) # converting to numpy array
z= linkage(df_norm,method='single',metric='euclidean')
plt.figure(figsize=(10,8))
plt.title('Hierarchial clustering dendogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    #leaf_rotation=6.,  # rotates the x axis labels
    #leaf_font_size=15.,  # font size for the x axis labels
)
plt.show()    
```

```python
p = np.array(df_norm) # converting to numpy array
z= linkage(df_norm,method='average',metric='euclidean')
plt.figure(figsize=(10,8))
plt.title('Hierarchial clustering dendogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    #leaf_rotation=6.,  # rotates the x axis labels
    #leaf_font_size=15.,  # font size for the x axis labels
)
plt.show()    
```

```python
p = np.array(df_norm)  # converting to numpy array
z= linkage(df_norm,method='complete',metric='euclidean')
plt.figure(figsize=(10,8))
plt.title('Hierarchial clustering dendogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    #leaf_rotation=6.,  # rotates the x axis labels
    #leaf_font_size=15.,  # font size for the x axis labels
)
plt.show()    
```

```python
p = np.array(df_norm) # converting to numpy array
z= linkage(crime_subset,method='single',metric='euclidean')
plt.figure(figsize=(10,8))
plt.title('Hierarchial clustering dendogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    #leaf_rotation=6.,  # rotates the x axis labels
    #leaf_font_size=15.,  # font size for the x axis labels
)
plt.show()    
```
